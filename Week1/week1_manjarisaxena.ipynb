### 1.3 Exceptions

Write `safe_divide(a, b)` that:
- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    try:
        if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
            raise TypeError("Inputs must be numeric")

        return a / b

    except ZeroDivisionError:
        return None

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None

try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"

print("1.3 passed ✓")
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[4, 1],
              [2, 3]], dtype=float)

# YOUR CODE HERE
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]

    left_side = A @ v
    right_side = lam * v

    print(f"\nEigenpair {i+1}:")
    print("A @ v =", left_side)
    print("λ * v =", right_side)

    assert np.allclose(left_side, right_side)

# Plot
fig, ax = plt.subplots()

for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]

    # Scale eigenvector by eigenvalue
    scaled_v = lam * v

    ax.quiver(
        0, 0,
        scaled_v[0], scaled_v[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        label=f'Eigenvector {i+1}'
    )

ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)

ax.set_xlabel('x')
ax.set_ylabel('y')

ax.set_title('Eigenvectors Scaled by Eigenvalues')

ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(0, color='black', linewidth=0.8)

ax.legend()

plt.grid(True)
plt.show()

print("4.3 passed ✓")
````

### Geometric Meaning of Eigenvalues and Eigenvectors

Eigenvectors are special vectors whose direction does not change when a matrix transformation is applied.  
Normally, a matrix can rotate, stretch, shrink, or flip vectors, but eigenvectors only change in magnitude while keeping the same direction.

Eigenvalues describe how much the corresponding eigenvectors are stretched or squished:

- If the eigenvalue is greater than 1, the vector stretches.
- If the eigenvalue is between 0 and 1, the vector shrinks (squishes).
- If the eigenvalue is negative, the vector also flips direction.

Geometrically, this means a matrix transformation scales vectors along certain special directions called eigenvectors. The amount of scaling in those directions is determined by the eigenvalues.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
import numpy as np

np.random.seed(42)

# Create random 4x3 matrix
X = np.random.randn(4, 3)

# YOUR CODE HERE
U, S, Vt = np.linalg.svd(X, full_matrices=False)

# Reconstruct X from SVD
X_reconstructed = U @ np.diag(S) @ Vt

# Rank-1 approximation
X_approx = S[0] * np.outer(U[:, 0], Vt[0, :])

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")

print(f"Singular values: {S}")

print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")

print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"

print("4.4 passed ✓")
````

### SVD → PCA Connection

In Singular Value Decomposition (SVD), a matrix X is decomposed into:

X = U S Vᵀ

where:

- U contains the left singular vectors.
- S contains the singular values, which measure the importance or strength of each component.
- Vᵀ contains the right singular vectors.

In PCA (Principal Component Analysis), the principal components are the directions of maximum variance in the data.

The matrix Vᵀ contains these principal component directions.  
Each row of Vᵀ represents a principal component axis.

The singular values in S indicate how much variance is captured by each principal component:
- Larger singular values correspond to directions with higher variance.
- Smaller singular values correspond to less important directions.

Thus, SVD provides the mathematical foundation for PCA and dimensionality reduction.

### 5.3 Error Metrics

Given these predictions and actuals, **implement from scratch** (no sklearn):
- MAE, MSE, RMSE
- R² Score
- Adjusted R² (assume 2 features)

````python
import numpy as np

y_true = np.array([3.0, 5.0, 2.5, 7.0, 4.5, 6.0, 1.5, 8.0])

y_pred = np.array([2.8, 5.2, 2.1, 7.5, 4.0, 6.3, 2.0, 7.8])

n, p = len(y_true), 2

# YOUR CODE HERE

# MAE
mae = np.mean(np.abs(y_true - y_pred))

# MSE
mse = np.mean((y_true - y_pred) ** 2)

# RMSE
rmse = np.sqrt(mse)

# R² Score
ss_res = np.sum((y_true - y_pred) ** 2)

ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)

r2 = 1 - (ss_res / ss_tot)

# Adjusted R²
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"MAE:        {mae:.4f}")

print(f"MSE:        {mse:.4f}")

print(f"RMSE:       {rmse:.4f}")

print(f"R²:         {r2:.4f}")

print(f"Adj. R²:    {adj_r2:.4f}")
````

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from scipy import stats
from statsmodels.tsa.stattools import adfuller
import numpy as np

np.random.seed(0)

# Generate samples
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
# YOUR CODE HERE
ks_s1 = stats.kstest(s1, 'norm')

ks_s2 = stats.kstest(s2, 'norm')

print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")

print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)

ts = 0.05 * t + np.random.normal(0, 1, 200)

# YOUR CODE HERE
adf_result = adfuller(ts)

print(f"\nADF statistic: {adf_result[0]:.4f}")

print(f"ADF p-value:   {adf_result[1]:.4f}")

print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
# YOUR CODE HERE
ts_diff = np.diff(ts)

adf_diff = adfuller(ts_diff)

print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

````python
import numpy as np
import matplotlib.pyplot as plt

def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """

    epsilon = 1e-10

    # Create bins based on expected distribution
    breakpoints = np.linspace(
        np.min(expected),
        np.max(expected),
        bins + 1
    )

    # Histogram counts
    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _ = np.histogram(actual, bins=breakpoints)

    # Convert to percentages
    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    # Avoid division by zero
    expected_pct = np.where(expected_pct == 0, epsilon, expected_pct)
    actual_pct = np.where(actual_pct == 0, epsilon, actual_pct)

    # PSI calculation
    psi = np.sum(
        (actual_pct - expected_pct) *
        np.log(actual_pct / expected_pct)
    )

    return psi


np.random.seed(1)

train_dist = np.random.normal(50, 10, 1000)    # training distribution

drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)

print(f"PSI: {psi_value:.4f}")

print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()

# YOUR CODE HERE
ax.hist(train_dist, bins=30, alpha=0.6, density=True, label='Training Distribution')

ax.hist(drift_dist, bins=30, alpha=0.6, density=True, label='Production Distribution')

ax.legend()

ax.set_title('Distribution Shift Visualization')

plt.tight_layout()

plt.show()
````

### Concept Drift vs Covariate Drift & PSI Thresholds (5.5)

- **Concept drift:**  
Concept drift occurs when the relationship between input features and the target variable changes over time, causing model performance to degrade.

- **Covariate drift:**  
Covariate drift occurs when the distribution of input features changes, even if the relationship with the target variable remains the same.

- **PSI < 0.1:**  
Indicates little or no significant distribution shift; the model is considered stable.

- **PSI 0.1–0.2:**  
Indicates a moderate or minor shift that should be monitored.

- **PSI > 0.2:**  
Indicates a major distribution shift and possible model degradation.

- **Retraining trigger example:**  
If PSI remains above 0.2 for several monitoring periods or model accuracy drops significantly, the model should be retrained using newer data.

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05

p_ham = 1 - p_spam

# P(free) — law of total probability
# YOUR CODE HERE
p_free = (
    p_free_given_spam * p_spam
    +
    p_free_given_ham * p_ham
)

# Bayes
# YOUR CODE HERE
p_spam_given_free = (
    p_free_given_spam * p_spam
) / p_free

print(f"P('free'): {p_free:.4f}")

print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")


def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):

    p_ham = 1 - prior_spam

    # Total probability of the word
    p_word = (
        p_word_given_spam * prior_spam
        +
        p_word_given_ham * p_ham
    )

    # Bayes theorem
    posterior = (
        p_word_given_spam * prior_spam
    ) / p_word

    return posterior


pred = naive_bayes_predict(0.30, 0.80, 0.05)

assert abs(pred - p_spam_given_free) < 1e-9

print(f"naive_bayes_predict: {pred:.4f}")

print("6.3 passed ✓")
````

| Evidence | P(word) | Overall probability of seeing the word “free” in any email |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(7)

# Population from exponential distribution
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30

# YOUR CODE HERE — array of 5000 sample means
sample_means = np.array([
    np.random.choice(population, sample_size).mean()
    for _ in range(n_samples)
])

pop_mean = population.mean()

pop_std = population.std()

# YOUR CODE HERE — predicted std of sample means
clt_std = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")

print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")

print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()

# YOUR CODE HERE — histogram of sample means
ax.hist(
    sample_means,
    bins=40,
    density=True,
    alpha=0.7,
    label='Sample Means'
)

# Theoretical normal overlay
x = np.linspace(
    sample_means.min(),
    sample_means.max(),
    300
)

ax.plot(
    x,
    stats.norm.pdf(x, loc=pop_mean, scale=clt_std),
    linewidth=2,
    label='Theoretical Normal'
)

ax.legend()

ax.set_title('CLT: Distribution of Sample Means')

plt.tight_layout()

plt.show()

# KS test
ks_result = stats.kstest(
    sample_means,
    'norm',
    args=(pop_mean, clt_std)
)

print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

### CLT Reflection (6.4)

The Central Limit Theorem states that as the sample size becomes sufficiently large, the distribution of sample means approaches a normal distribution, even if the original population distribution is not normal.

This result is important in statistics and machine learning because it allows us to use normal-distribution-based methods for estimation, confidence intervals, and hypothesis testing. The CLT justifies the normality assumption used in many classical statistical tests such as the t-test and z-test.
