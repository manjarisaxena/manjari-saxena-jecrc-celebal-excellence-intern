# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION (IMPLEMENTED)

import logging

logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
logger = logging.getLogger("agent")


def agent(query: str):
    logger.info(f"Received query: {query}")
    query_lower = query.lower()

    try:
        # 🧮 Route 1: Calculator
        if "calculate" in query_lower:
            # Extract the mathematical expression from the query
            expression = query_lower.replace("calculate", "").strip()
            expression = expression.rstrip("?.! ")

            if not expression:
                logger.warning("No expression found for calculation")
                return {
                    "type": "error",
                    "result": "No mathematical expression found to calculate"
                }

            logger.info(f"Routing to Calculator Tool with expression: {expression}")
            calc_result = calculator(expression)

            if calc_result == "Error in calculation":
                return {
                    "type": "error",
                    "result": calc_result
                }

            return {
                "type": "calculation",
                "result": calc_result
            }

        # 🔑 Route 2: Keyword Extractor
        elif "keywords" in query_lower:
            # Remove trigger phrase, keep the actual text to analyze
            text = query
            for phrase in ["extract keywords from", "keywords from", "extract keywords"]:
                text = text.lower().replace(phrase, "")
            text = text.strip()

            if not text:
                logger.warning("No text found for keyword extraction")
                return {
                    "type": "error",
                    "result": "No text found to extract keywords from"
                }

            logger.info(f"Routing to Keyword Extractor Tool with text: {text}")
            keywords = extract_keywords(text)

            return {
                "type": "keywords",
                "result": keywords
            }

        # 💬 Route 3: General Response
        else:
            logger.info("Routing to General Response")
            return {
                "type": "general",
                "result": f"I received your query: '{query}'. This is a general response since no specific tool was matched."
            }

    except Exception as e:
        logger.error(f"Error while processing query: {e}")
        return {
            "type": "error",
            "result": f"Something went wrong while processing your query: {e}"
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```
